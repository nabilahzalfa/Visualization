```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = '/content/drive/MyDrive/Colab Notebooks/MSIB/Pengenalan Python untuk Data Science/PDA_DataSource/Data Visualization/percent-bachelors-degrees-women-usa.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

recall challange:
- buat line plot dari physical science dan computer science terhadap year
- tampilkan legendanya
- gunakan line plot dan matplotlib

```python
plt.plot('Year', 'Computer Science', data=df)
plt.plot('Year', 'Physical Sciences', data=df)
plt.legend(loc='lower center')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.show()
```

```python
# tips
# definisikan dulu kolom - kolom yang akan digunakan
year = df['Year']
physical_science = df['Physical Sciences']
computer_science = df['Computer Science']
Psychology = df['Psychology']
business = df['Business']
```

```python
plt.plot(year, computer_science); # == plt.plot(df['Year'], df['Computer Science'])
```

## Menggunakan subplot

```python
plt.figure(figsize=(10,6))
plt.plot(year, business)
plt.show()
```

```python
# satu visual
plt,plt.subplots(1, 1, figsize=(5,3))
plt.plot(year, business)
plt.show()
```

```python
# variable subplot
# fig = mengacu pada figsize
# ax = merujuk pada subplotnya
fig, ax = plt.subplots(1, 1, figsize=(5,3))
ax.plot(year, business)
sns.lineplot(x=year, y=physical_science, ax=ax)
```

```python
# multiple visual
fig, ax = (plt.subplots(2,1, figsize=(8,5)))
ax[0].plot(year, business)
sns.lineplot(x=year, y=physical_science, ax=ax[1], color='red')
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,6))
ax[0].plot(year, Psychology)
sns.lineplot(x=year, y=computer_science, ax=ax[1], color='green');
ax[0].set_title('Psychology over Year', fontsize=16)
ax[1].set_title('Computer Science over Year', fontsize= 16)
fig.tight_layout(pad=1)
```

## Label

```python
plt.plot(year, business, color='#5f5faf', linestyle='dashdot', linewidth=3)
plt.title('Business over Year', fontsize=16)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Percentage Business Degree', fontsize=10)
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,6))
ax[0].plot(year, Psychology)
ax[1].plot(year, computer_science, color='green');
ax[0].set_title('Psychology over Year', fontsize=16)
ax[1].set_title('Computer Science over Year', fontsize= 16)
fig.tight_layout(pad=5)

#memberikan label
ax[0].set_xlabel('Year', fontsize=10)
ax[1].set_xlabel('Year', fontsize=10)
fig.text(0.05, 0.4, 'Percentage Bachelor', rotation='vertical', fontsize=10) # memberikan text bebas
plt.show()
```

```python
plt.plot('Year', 'Computer Science', data=df)
plt.plot('Year', 'Physical Sciences', data=df)
plt.legend(loc='upper center')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.xlim(1985, 1995)
plt.ylim(25, 40)
plt.xticks(list(range(1985, 1996, 5)))
plt.yticks(list(range(25, 41, 5)))
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,6))
ax[0].plot(year, Psychology, label='Psy')
ax[0].plot(year, computer_science, color='green', label='cs');
ax[0].set_title('Psychology over Year', fontsize=16)
ax[1].set_title('Computer Science over Year', fontsize= 16)
fig.tight_layout(pad=5)

#memberikan label
ax[0].set_xlabel('Year', fontsize=10)
ax[1].set_xlabel('Year', fontsize=10)
fig.text(0.05, 0.4, 'Percentage Bachelor', rotation='vertical', fontsize=10) # memberikan text bebas

# menambahkan legend
ax[0].legend(loc='upper center')
plt.show()
```
